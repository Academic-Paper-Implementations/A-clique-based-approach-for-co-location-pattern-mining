## Code tạo tập F từ tập S nhằm xác định thứ tự của các feature

```python
import pandas as pd

# Đọc file S.csv
df = pd.read_csv('data/S.csv')

# Lấy các giá trị unique của cột Feature và sắp xếp theo thứ tự tăng dần
unique_features = sorted(df['Feature'].unique())

# Tạo DataFrame mới với cột Feature
df_output = pd.DataFrame({'Feature': unique_features})

# Lưu ra file F.csv
df_output.to_csv('data/F.csv', index=False)

print("Đã tạo file F.csv thành công!")
print("\nDanh sách các feature:")
print(df_output)
```
